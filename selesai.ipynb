```python
# Import libraries
import pandas as pd
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from google.colab import drive
import joblib
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset dari Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Path ke file dataset Anda
data_path = '/content/drive/MyDrive/UAS-NLP-kelompok-04/main1.csv'

data = pd.read_csv(data_path)
```

```python
# Membaca dataset
import pandas as pd
file_path = '/content/drive/MyDrive/UAS-NLP-kelompok-04/main1.csv'  # Define file_path here with your data path
dataset = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Menyiapkan data untuk pelatihan
input_patterns = dataset['pattern']  # Pola input
responses = dataset['response']      # Respons/tag yang sesuai

# Membagi data menjadi set pelatihan dan pengujian
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    input_patterns, responses, test_size=0.2, random_state=42
)
```

```python
# Create and train the model
model = make_pipeline(CountVectorizer(), SVC(kernel='linear'))
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
from joblib import dump

# Filepath untuk menyimpan model dan vectorizer
model_filename = '/content/drive/MyDrive/UAS-NLP-kelompok-04/chatbot_model.pkl'  # Sesuaikan dengan path
vectorizer_filename = '/content/drive/MyDrive/UAS-NLP-kelompok-04/count_vectorizer.pkl'  # Path untuk CountVectorizer

# Simpan pipeline model yang sudah mencakup vectorizer
dump(model, model_filename)
# Simpan hanya CountVectorizer secara terpisah (jika diperlukan)
dump(model.named_steps['countvectorizer'], vectorizer_filename)

# Pesan konfirmasi
print(f"\033[91mModel berhasil tersimpan di: {model_filename}\033[0m")  # Warna merah
print(f"\033[91mCountVectorizer berhasil tersimpan di: {vectorizer_filename}\033[0m")  # Warna merah
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
# Print evaluation results
print("\033[92mHasil Evaluasi Model:\033[0m")
print(f"Akurasi: {accuracy:.2f}")
print("\n\033[94mLaporan Klasifikasi:\033[0m")
print(report)
```

```python
# Plot confusion matrix with a different style
plt.figure(figsize=(50, 70))
sns.set_theme(style="whitegrid")
sns.heatmap(conf_matrix, annot=True, fmt='.1f', cmap='coolwarm', linewidths=10,
            cbar=True, square=True, xticklabels=model.classes_, yticklabels=model.classes_)
plt.ylabel('Label Pertanyaan')
plt.xlabel('Label Jawaban')
plt.title('Matriks Kebingungan')
plt.tight_layout()
```
